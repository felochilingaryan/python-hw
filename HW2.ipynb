```python
numbers=input("some numbers separated by commas")
def average(numbers):
    return float(sum(numbers))/len(numbers)
print average(numbers)
```

```python
numbers=input("some numbers separated by commas")
argument=input("type some argument")
def average(numbers):
    return float(sum(numbers[-(argument):]))/float(int(argument))
print average(numbers)
```
